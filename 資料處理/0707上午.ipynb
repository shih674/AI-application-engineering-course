```python
import csv
import codecs

with codecs.open('.\\第14任總統副總統選舉.csv', encoding='big5') as f:
    r = csv.reader(f) #上下二選一，都可以執行
#    r = list(csv.reader(f)) #上下二選一，都可以執行
    for d in r:
        print(d)
```

## 注意!! 
### 用來裝csv內容的"變數f" 在csv.reader解析之後內容會被清空! 所以記得拿其他變數來裝內容

```python
import requests
import io

url = 'https://quality.data.gov.tw/dq_download_csv.php?nid=94202&md5_url=71244f8decaa59c2f234717abadef892'
r1 = requests.get(url)
r1.encoding = 'utf8'
print(r1.text)
f = io.StringIO(r1.text)
r = list(csv.reader(f))
for d in r:
    print(d[0], d[1], d[2])
```

```python
import json
import io

url = 'https://quality.data.gov.tw/dq_download_json.php?nid=13348&md5_url=6897aa8e5fed5b989d853e88613e4f54'
json_file = requests.get(url)
json_content = json.loads(json_file.text)
print('文件內容示範列: ',json_content[0])

for ele in json_content:
    print(ele['核定年月'], ele['給付種類'], ele['受理件數'])
```
