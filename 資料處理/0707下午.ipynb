```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('https://www.taiwan.net.tw/m1.aspx',
                     params={
                         'sNo':'0001001'
                     })
bs = BeautifulSoup(source.text, 'html.parser')
bs
```

```python
find_something = bs.find_all('a',{'class':'columnBlock-title','target':'_self'})
find_something
```

```python
count = 1
for ele in find_something:
    print(count, ele.text)
    count += 1
```

```python
find_something1 = bs.find_all('span',{'class':'date'})
for ele in find_something1:
    print(ele.text)
```

```python
# 一層一層爬
find_something2 = bs.find_all('div', {'class':'columnBlock-info'})
for ele in find_something2:
    date  = ele.find('span', {'class':'date'}).text
    link  = ele.find('a',{'class':'columnBlock-title','target':'_self'})
    title = link.text
    url   = link.attrs['href']
    print(date,'\n',title,'\n',url,'\n')
```
