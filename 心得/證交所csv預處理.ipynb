## 預先處理，把從證交所抓回來的csv轉成pandas能讀取的格式

### 資料來源網址: https://www.twse.com.tw/zh/page/trading/fund/T86.html

#### step1. 從本地端讀出檔案
            * 原先的檔案編碼為 'ANSI'
#### step2. 以"列"為單位，切開整份資料
#### step3. 找出資料的first index與final index，用這兩個index把資料倒出來
            * index = 0 : 109年06月24日 三大法人買賣超日報
                我要拿來當'檔案名稱'
            * 最後幾列是"文件的說明"，對我來說是雜訊，一樣不要。 因為"說明"是固定模式→固定列數
                找出"說明"佔幾列，從後面算回來刪掉
#### step4. 排成能存檔的str
            * 資料型別List是無法直接用with open語法做儲存的，因此要改成資料型別String同時具有csv的格式
            * csv格式: "\n"→分隔列  ","→ 分隔欄位
#### step5. 檔名轉化成英文
            * pandas.read_csv 無法讀取中文檔名 = ="
#### step6. 整理好的表格儲存起來，設定編碼為'utf-8'

```python
# step1. 從本地端讀出檔案
fileName = 'T8600.csv'
Encoding = 'ANSI'
with open(fileName, encoding=Encoding) as f:
    raw_data = f.read()
print(raw_data)
```

```python
# step2. 以"列"為單位，切開整份資料

raw_data_list = raw_data.split('\n')

# 印出來檢查
for ele in raw_data_list:
    print(ele,'\n')
```

```python
# step3-1. 找出我要的資料的index
end_of_data = len(raw_data_list) - 10

# 印出來檢查
for i in range(1,end_of_data):
    print(raw_data_list[i],'\n')
```

```python
# step3-2. 用剛剛找到的index倒出我需要的資料
save_name = raw_data_list[0].replace('"','')
print(save_name)

raw_data_clean = raw_data_list[1:len(raw_data_list)-10]
raw_data_clean
```

```python
# step4: 排成能存檔的str
save_data = '\n'.join(raw_data_clean)
save_data
```

```python
# step5. 檔名轉化成英文

# 處理詞 "三大法人買賣超日報"
save_name = save_name.replace('三大法人買賣超日報','Daily Trading Details of Foreign and Other Investors')

# 處理詞 "年","月","日"
save_name = save_name.replace('年','-')
save_name = save_name.replace('月','-')
save_name = save_name.replace('日','')

# 加上副檔名
save_name = save_name + '.csv'

save_name
```

```python
# step6. 整理好的表格儲存起來，設定編碼為'utf-8'
Encoding = 'utf-8'
with open(save_name,'a',encoding=Encoding) as f:
    f.write(save_data)
```

## 肉眼觀察表格內的資訊，找出哪些地方需要調整?
    * 證券代號: 只要出現英文，都會出現 = 和 " "→ 要去掉
    * 數字格式: 數字以會計常見的,呈現 → 去掉','
    * 外資自營商的三個欄位都沒帶值 → 去掉
    * Unnamed: 19 :內容都是NaN → 去掉

```python
import pandas as pd
```

```python
# pd.read_csv會有"無法讀取中文檔名"的情況!
filename = save_name
Encoding = 'utf-8'
df = pd.read_csv(filename, encoding=Encoding)
df
```

```python
# 調正正確代號的錯誤
for i in range(len(df['證券代號'])):
    if '=' in df['證券代號'][i]:
        s = df['證券代號'][i]
        s = s.replace('=', '')
        df['證券代號'][i] = s.replace('"', '')
        
    print(i,df['證券代號'][i])
```

```python
#再次檢查表格
df
```

```python
# 數字格式: 數字以會計常見的,呈現 → 去掉','

def transfer(string):
    if string != 0:
        string = string.replace(',','')
    return string

# 需要調整內容的"行"
title_list = ['外陸資買進股數(不含外資自營商)', '外陸資賣出股數(不含外資自營商)', '外陸資買賣超股數(不含外資自營商)', 
              '外資自營商買進股數', '外資自營商賣出股數', '外資自營商買賣超股數',
              '投信買進股數', '投信賣出股數', '投信買賣超股數', 
              '自營商買進股數(自行買賣)', '自營商賣出股數(自行買賣)', '自營商買賣超股數(自行買賣)',
              '自營商買進股數(避險)', '自營商賣出股數(避險)', '自營商買賣超股數(避險)', '三大法人買賣超股數']

for ele in title_list:
    df[ele] = df[ele].apply(transfer)
    
df
```

```python
# 外資自營商的三個欄位都沒帶值 → 去掉
df = df.drop(['外資自營商買進股數', '外資自營商賣出股數', '外資自營商買賣超股數'], axis = 1)

# Unnamed: 19 :內容都是NaN → 去掉
df = df.drop('Unnamed: 19', axis = 1)
df
```

```python
# 整理好的表格儲存起來，設定編碼為'utf-8'
df.to_csv('0624.csv', encoding='utf-8')
```
